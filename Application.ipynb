```python
import warnings
warnings.filterwarnings("ignore", message="A matching Triton is not available")

import torch
import torch.nn.functional as F
from torch import autocast
import numpy as np
from PIL import Image
import psutil
import os
import time
import gc
from typing import Optional, Tuple, List
from datetime import datetime
from importlib.metadata import version

from diffusers import (
    StableDiffusionPipeline,
    EulerAncestralDiscreteScheduler,
    EulerDiscreteScheduler,
    DPMSolverMultistepScheduler,
    DDIMScheduler,
    LMSDiscreteScheduler
)
import gradio as gr
```

```python
class StableDiffusionGenerator:
    def __init__(self, model_id: str = "runwayml/stable-diffusion-v1-5", device: str = "auto"):
        try:
            self.device = self._setup_device(device)
            self.dtype = torch.float16 if self.device.type == "cuda" else torch.float32

            print(f"🚀 Initializing Stable Diffusion on {self.device}")
            print(f"📊 Using precision: {self.dtype}")

            torch_version = version("torch")
            diffusers_version = version("diffusers")
            print(f"📦 PyTorch version: {torch_version}")
            print(f"📦 Diffusers version: {diffusers_version}")

            self.pipe = self._load_pipeline(model_id)
            self.current_scheduler = "euler_a"
            self.schedulers = {
                "euler_a": ("Euler Ancestral", "Fast, good for creative images"),
                "euler": ("Euler", "Deterministic, consistent results"),
                "ddim": ("DDIM", "Classic, good quality, slower"),
                "dpm_solver": ("DPM Solver", "High quality, efficient"),
                "lms": ("LMS", "Linear multistep, stable")
            }
            print("✅ Stable Diffusion Generator Ready!")
            print(f"📝 Available Schedulers: {list(self.schedulers.keys())}")
        except Exception as e:
            print(f"❌ Initialization Error: {str(e)}")
            print("Please ensure Visual C++ Redistributable 2015-2022 is installed")
            raise

    def _setup_device(self, device: str) -> torch.device:
        if device == "auto":
            if torch.cuda.is_available():
                device = "cuda"
                print(f"🎯 GPU Detected: {torch.cuda.get_device_name(0)}")
                print(f"💾 VRAM: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f}GB")
            else:
                device = "cpu"
                print("💻 Using CPU (GPU not available)")
        return torch.device(device)

    def _load_pipeline(self, model_id: str) -> StableDiffusionPipeline:
        try:
            pipe = StableDiffusionPipeline.from_pretrained(
                model_id,
                torch_dtype=self.dtype,
                safety_checker=None,
                requires_safety_checker=False,
            )
            print("🔧 Applying Memory Optimizations...")
            pipe.enable_attention_slicing()
            print("  ✓ Attention Slicing: Enabled")
            pipe.enable_vae_slicing()
            print("  ✓ VAE Slicing: Enabled")
            try:
                pipe.enable_xformers_memory_efficient_attention()
                print("  ✓ XFormers Attention: Enabled")
            except Exception as e:
                print(f"  ⚠ XFormers: Not available ({e})")
            if self.device.type == "cuda":
                try:
                    pipe = pipe.to(self.device)
                    print("  ✓ Full GPU Loading: Success")
                except RuntimeError as e:
                    print("  ⚠ GPU Memory Limited: Using CPU Offload")
                    pipe.enable_model_cpu_offload()
            else:
                pipe.enable_sequential_cpu_offload()
                print("  ✓ CPU Sequential Offload: Enabled")
            return pipe
        except Exception as e:
            raise RuntimeError(f"Failed to load model: {e}")

    def set_scheduler(self, scheduler_name: str) -> bool:
        if scheduler_name not in self.schedulers:
            print(f"❌ Unknown scheduler: {scheduler_name}")
            return False
        if scheduler_name == self.current_scheduler:
            return True
        scheduler_map = {
            "euler_a": EulerAncestralDiscreteScheduler,
            "euler": EulerDiscreteScheduler,
            "ddim": DDIMScheduler,
            "dpm_solver": DPMSolverMultistepScheduler,
            "lms": LMSDiscreteScheduler
        }
        try:
            scheduler_class = scheduler_map[scheduler_name]
            self.pipe.scheduler = scheduler_class.from_config(self.pipe.scheduler.config)
            self.current_scheduler = scheduler_name
            name, desc = self.schedulers[scheduler_name]
            print(f"🔄 Scheduler Changed: {name} ({desc})")
            return True
        except Exception as e:
            print(f"❌ Scheduler Error: {e}")
            return False

    def generate_image(
        self,
        prompt: str,
        negative_prompt: str = "",
        width: int = 512,
        height: int = 512,
        num_inference_steps: int = 20,
        guidance_scale: float = 7.5,
        seed: Optional[int] = None,
        scheduler: str = "euler_a"
    ) -> Tuple[Image.Image, dict]:
        if not prompt.strip():
            raise ValueError("Prompt cannot be empty")
        self.set_scheduler(scheduler)
        if seed is None:
            seed = torch.randint(0, 2**32, (1,)).item()
        generator = torch.Generator(device=self.device)
        generator.manual_seed(seed)
        width = (width // 8) * 8
        height = (height // 8) * 8
        print(f"🎨 Generating: '{prompt[:50]}...'")
        print(f"📏 Size: {width}x{height}, Steps: {num_inference_steps}, CFG: {guidance_scale}")
        print(f"🎲 Seed: {seed}, Scheduler: {scheduler}")
        start_time = time.time()
        try:
            with torch.inference_mode():
                if self.device.type == "cuda" and self.dtype == torch.float16:
                    with autocast(self.device.type):
                        result = self.pipe(
                            prompt=prompt,
                            negative_prompt=negative_prompt if negative_prompt else None,
                            width=width,
                            height=height,
                            num_inference_steps=num_inference_steps,
                            guidance_scale=guidance_scale,
                            generator=generator
                        )
                else:
                    result = self.pipe(
                        prompt=prompt,
                        negative_prompt=negative_prompt if negative_prompt else None,
                        width=width,
                        height=height,
                        num_inference_steps=num_inference_steps,
                        guidance_scale=guidance_scale,
                        generator=generator
                    )
            generation_time = time.time() - start_time
            metadata = {
                "prompt": prompt,
                "negative_prompt": negative_prompt,
                "width": width,
                "height": height,
                "steps": num_inference_steps,
                "guidance_scale": guidance_scale,
                "scheduler": scheduler,
                "seed": seed,
                "generation_time": round(generation_time, 2),
                "device": str(self.device),
                "dtype": str(self.dtype)
            }
            print(f"✅ Generated in {generation_time:.2f}s")
            return result.images[0], metadata
        except torch.cuda.OutOfMemoryError:
            self._cleanup_memory()
            raise RuntimeError(
                "GPU Out of Memory! Try: reducing image size, fewer steps, "
                "or use CPU mode. Current settings may be too demanding."
            )
        except Exception as e:
            raise RuntimeError(f"Generation failed: {str(e)}")
        finally:
            self._cleanup_memory()

    def _cleanup_memory(self):
        gc.collect()
        if self.device.type == "cuda":
            torch.cuda.empty_cache()

    def get_memory_usage(self) -> dict:
        memory_info = {}
        if self.device.type == "cuda":
            memory_info = {
                "allocated_gb": torch.cuda.memory_allocated() / 1024**3,
                "reserved_gb": torch.cuda.memory_reserved() / 1024**3,
                "max_allocated_gb": torch.cuda.max_memory_allocated() / 1024**3,
                "total_gb": torch.cuda.get_device_properties(0).total_memory / 1024**3
            }
        else:
            memory_info = {"device": "cpu", "note": "CPU memory tracking not available"}
        return memory_info

    def save_image(self, image: Image.Image, metadata: dict, output_dir: str = "outputs") -> str:
        os.makedirs(output_dir, exist_ok=True)
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"sd_gen_{timestamp}_s{metadata['seed']}_{metadata['width']}x{metadata['height']}.png"
        filepath = os.path.join(output_dir, filename)
        image.save(filepath)
        metadata_file = filepath.replace('.png', '_metadata.txt')
        with open(metadata_file, 'w') as f:
            f.write("Stable Diffusion Generation Metadata\n")
            f.write("=" * 40 + "\n")
            for key, value in metadata.items():
                f.write(f"{key}: {value}\n")
        print(f"💾 Saved: {filepath}")
        return filepath
```

```python
class StableDiffusionUI:

    def __init__(self):
        self.generator = None
        self.gallery_images = []
        self.generation_history = []

    def initialize_generator(self, model_choice: str, device_choice: str) -> str:
        try:
            model_map = {
                "Stable Diffusion 1.5 (Recommended)": "runwayml/stable-diffusion-v1-5",
                "Stable Diffusion 2.1": "stabilityai/stable-diffusion-2-1",
                "Realistic Vision (RealVisXL)": "SG161222/RealVisXL_V4.0"
            }
            device_map = {
                "Auto (Recommended)": "auto",
                "GPU (CUDA)": "cuda",
                "CPU (Slower)": "cpu"
            }
            model_id = model_map.get(model_choice, "runwayml/stable-diffusion-v1-5")
            device = device_map.get(device_choice, "auto")
            self.generator = StableDiffusionGenerator(model_id=model_id, device=device)
            memory_info = self.generator.get_memory_usage()
            memory_text = f"Memory Usage: {memory_info}" if memory_info else "Ready!"
            return f"✅ Model loaded successfully!\n{memory_text}"
        except Exception as e:
            return f"❌ Initialization failed: {str(e)}"

    def generate_image(
        self,
        prompt: str,
        negative_prompt: str,
        width: int,
        height: int,
        steps: int,
        guidance: float,
        scheduler: str,
        seed: int,
        save_image: bool
    ):
        if self.generator is None:
            return None, "❌ Please initialize the model first!", ""
        if not prompt.strip():
            return None, "❌ Please enter a prompt!", ""
        try:
            seed = None if seed == -1 else int(seed)
            image, metadata = self.generator.generate_image(
                prompt=prompt,
                negative_prompt=negative_prompt,
                width=width,
                height=height,
                num_inference_steps=steps,
                guidance_scale=guidance,
                scheduler=scheduler,
                seed=seed
            )
            info_text = self._format_generation_info(metadata)
            saved_path = ""
            if save_image:
                saved_path = self.generator.save_image(image, metadata)
            self.generation_history.append(metadata)
            self.gallery_images.append(image)
            if len(self.gallery_images) > 10:
                self.gallery_images = self.gallery_images[-10:]
                self.generation_history = self.generation_history[-10:]
            return image, info_text, saved_path
        except Exception as e:
            return None, f"❌ Generation failed: {str(e)}", ""

    def _format_generation_info(self, metadata: dict) -> str:
        return f"""
✅ Generation Complete!

🎯 Parameters Used:
• Prompt: {metadata['prompt'][:100]}{'...' if len(metadata['prompt']) > 100 else ''}
• Size: {metadata['width']} × {metadata['height']}
• Steps: {metadata['steps']}
• Guidance Scale: {metadata['guidance_scale']}
• Scheduler: {metadata['scheduler']}
• Seed: {metadata['seed']}

⏱️ Performance:
• Generation Time: {metadata['generation_time']}s
• Device: {metadata['device']}
• Precision: {metadata['dtype']}
"""

    def get_example_prompts(self):
        return [
            ["a futuristic cyberpunk city with neon lights", "blurry, low quality"],
            ["portrait of a fantasy queen, cinematic lighting", "ugly, deformed"],
            ["minimal luxury modern interior", "dark, messy"],
            ["cute anime girl in sakura garden", "realistic, scary"],
            ["luxury sports car in rainy tokyo street", "low resolution"]
        ]

    def show_scheduler_info(self, scheduler: str):
        scheduler_info = {
            "euler_a": "Fast & creative generation",
            "euler": "Deterministic and stable",
            "ddim": "Classic quality scheduler",
            "dpm_solver": "Efficient premium quality",
            "lms": "Very stable outputs"
        }
        return scheduler_info.get(scheduler, "Scheduler information unavailable")

    def get_memory_info(self):
        if self.generator is None:
            return "Model not loaded"
        try:
            memory_info = self.generator.get_memory_usage()
            if 'allocated_gb' in memory_info:
                return f"""
GPU Memory Usage

• Allocated: {memory_info['allocated_gb']:.2f} GB
• Reserved: {memory_info['reserved_gb']:.2f} GB
• Total Available: {memory_info['total_gb']:.2f} GB
• Usage: {(memory_info['allocated_gb']/memory_info['total_gb']*100):.1f}%
"""
            else:
                return "CPU mode - memory tracking unavailable"
        except:
            return "Memory info unavailable"

    def create_interface(self):

        custom_css = """

        @import url('https://fonts.googleapis.com/css2?family=Syne:wght@400;600;700;800&family=DM+Sans:opsz,wght@9..40,300;9..40,400;9..40,500;9..40,600&display=swap');

        /* =============================================
           BASE RESET
        ============================================= */
        *, *::before, *::after { box-sizing: border-box; }

        html { background: #070712 !important; }

        body {
            background: #070712 !important;
            font-family: 'DM Sans', sans-serif !important;
            color: #e2e8f0 !important;
            margin: 0 !important;
            padding: 0 !important;
            min-height: 100vh !important;
        }

        body::before {
            content: '';
            position: fixed;
            inset: 0;
            background:
                radial-gradient(ellipse 90% 60% at 10% 0%,  rgba(109,40,217,0.16) 0%, transparent 55%),
                radial-gradient(ellipse 70% 50% at 90% 90%, rgba(219,39,119,0.11) 0%, transparent 55%);
            z-index: -1;
            pointer-events: none;
        }

        /* =============================================
           GRADIO CONTAINER
        ============================================= */
        .gradio-container {
            width: 100vw !important;
            max-width: 100vw !important;
            min-height: 100vh !important;
            padding: 0 !important;
            margin: 0 !important;
            background: transparent !important;
        }

        .gradio-container > *,
        .main, .contain, .wrap, .gap,
        .form, .gr-form, .gr-box, .gr-padded,
        .panel, .block, fieldset,
        .gr-group, .gr-column, .gr-row {
            background: transparent !important;
        }

        /* =============================================
           HIDE GRADIO FOOTER BAR ONLY
           Target Gradio's own footer, NOT our .site-footer
        ============================================= */
        .gradio-container ~ footer,
        footer.svelte-1ax1toq,
        footer.svelte-mpyp5e,
        div.svelte-1ax1toq,
        .show-api,
        .built-with,
        .api-docs,
        a[href*="gradio.app"],
        a[href*="huggingface.co/spaces"],
        button[aria-label="Settings"],
        button[title="Settings"],
        .settings-button,
        .svelte-1lnpkp2 {
            display: none !important;
            visibility: hidden !important;
            height: 0 !important;
            overflow: hidden !important;
            pointer-events: none !important;
        }

        /* =============================================
           NAVBAR
        ============================================= */
        .site-nav {
            width: 100%;
            display: flex;
            align-items: center;
            justify-content: space-between;
            padding: 0 52px;
            height: 64px;
            background: rgba(7,7,18,0.92) !important;
            backdrop-filter: blur(28px);
            -webkit-backdrop-filter: blur(28px);
            border-bottom: 1px solid rgba(255,255,255,0.06);
            position: sticky;
            top: 0;
            z-index: 999;
        }

        .nav-logo {
            font-family: 'Syne', sans-serif !important;
            font-size: 1.2rem;
            font-weight: 800;
            letter-spacing: -0.02em;
            background: linear-gradient(90deg, #c084fc, #f472b6);
            -webkit-background-clip: text;
            -webkit-text-fill-color: transparent;
        }

        .nav-pill {
            font-size: 11px;
            font-weight: 500;
            padding: 5px 14px;
            border-radius: 999px;
            border: 1px solid rgba(192,132,252,0.25);
            color: #c084fc;
            letter-spacing: 0.06em;
            text-transform: uppercase;
        }

        /* =============================================
           HERO
        ============================================= */
        .hero-section {
            width: 100%;
            text-align: center;
            padding: 80px 40px 60px;
        }

        .hero-eyebrow {
            display: inline-flex;
            align-items: center;
            gap: 8px;
            background: rgba(192,132,252,0.08);
            border: 1px solid rgba(192,132,252,0.2);
            border-radius: 999px;
            padding: 6px 18px;
            font-size: 11.5px;
            font-weight: 500;
            color: #c084fc;
            letter-spacing: 0.07em;
            text-transform: uppercase;
            margin-bottom: 28px;
        }

        .hero-dot {
            width: 6px; height: 6px;
            border-radius: 50%;
            background: #c084fc;
            animation: blink 2s ease-in-out infinite;
        }

        @keyframes blink {
            0%,100% { opacity: 1; }
            50%      { opacity: 0.25; }
        }

        .hero-title {
            font-family: 'Syne', sans-serif !important;
            font-size: clamp(2.8rem, 5.5vw, 5.2rem);
            font-weight: 800;
            letter-spacing: -0.04em;
            line-height: 1.02;
            color: #ffffff !important;
            margin-bottom: 18px;
        }

        .hero-title .grad {
            background: linear-gradient(120deg, #a78bfa 0%, #f472b6 55%, #fb923c 100%);
            -webkit-background-clip: text;
            -webkit-text-fill-color: transparent;
        }

        .hero-sub {
            color: rgba(255,255,255,0.35);
            font-size: clamp(0.9rem, 1.4vw, 1.05rem);
            max-width: 480px;
            margin: 0 auto 44px;
            line-height: 1.75;
        }

        .hero-stats {
            display: flex;
            justify-content: center;
            gap: 52px;
            flex-wrap: wrap;
        }

        .stat-item { text-align: center; }

        .stat-num {
            font-family: 'Syne', sans-serif !important;
            font-size: 1.65rem;
            font-weight: 700;
            color: #fff;
            display: block;
            line-height: 1;
            margin-bottom: 4px;
        }

        .stat-lbl {
            font-size: 10.5px;
            color: rgba(255,255,255,0.25);
            letter-spacing: 0.1em;
            text-transform: uppercase;
        }

        /* =============================================
           DIVIDER
        ============================================= */
        .divider {
            width: 100%;
            height: 1px;
            background: linear-gradient(90deg,
                transparent 0%,
                rgba(192,132,252,0.22) 25%,
                rgba(192,132,252,0.22) 75%,
                transparent 100%
            );
        }

        /* =============================================
           APP BODY
        ============================================= */
        .app-body {
            width: 100% !important;
            padding: 36px 40px 48px !important;
        }

        /* =============================================
           CARDS
        ============================================= */
        .modern-card {
            background: rgba(255,255,255,0.032) !important;
            border-radius: 22px !important;
            border: 1px solid rgba(255,255,255,0.07) !important;
            padding: 26px !important;
            transition: border-color 0.25s ease !important;
        }

        .modern-card:hover {
            border-color: rgba(192,132,252,0.2) !important;
        }

        .output-img-card {
            background: rgba(255,255,255,0.032) !important;
            border-radius: 22px !important;
            border: 1px solid rgba(255,255,255,0.07) !important;
            padding: 16px !important;
            transition: border-color 0.25s ease !important;
        }

        .output-img-card:hover {
            border-color: rgba(192,132,252,0.2) !important;
        }

        /* =============================================
           SIDEBAR
        ============================================= */
        .sidebar {
            position: sticky !important;
            top: 80px !important;
        }

        /* =============================================
           BUTTONS
        ============================================= */
        button {
            font-family: 'DM Sans', sans-serif !important;
            border-radius: 13px !important;
            border: none !important;
            font-weight: 500 !important;
            font-size: 14px !important;
            letter-spacing: 0.01em !important;
            transition: all 0.2s ease !important;
            height: 50px !important;
            cursor: pointer !important;
        }

        button:active { transform: scale(0.975) !important; }

        button.primary {
            background: linear-gradient(135deg, #7c3aed 0%, #c026d3 100%) !important;
            color: #fff !important;
            box-shadow: 0 1px 0 rgba(255,255,255,0.12) inset,
                        0 8px 24px rgba(124,58,237,0.32) !important;
        }

        button.primary:hover {
            background: linear-gradient(135deg, #6d28d9 0%, #a21caf 100%) !important;
            box-shadow: 0 1px 0 rgba(255,255,255,0.12) inset,
                        0 12px 32px rgba(124,58,237,0.48) !important;
            transform: translateY(-1px) !important;
        }

        button:not(.primary) {
            background: rgba(255,255,255,0.05) !important;
            color: rgba(255,255,255,0.6) !important;
            border: 1px solid rgba(255,255,255,0.09) !important;
        }

        button:not(.primary):hover {
            background: rgba(255,255,255,0.09) !important;
            color: rgba(255,255,255,0.88) !important;
        }

        /* =============================================
           INPUTS
        ============================================= */
        textarea,
        input[type="text"],
        input[type="number"],
        input[type="search"] {
            font-family: 'DM Sans', sans-serif !important;
            border-radius: 13px !important;
            border: 1px solid rgba(255,255,255,0.09) !important;
            background: rgba(255,255,255,0.045) !important;
            color: #f0f4ff !important;
            font-size: 14px !important;
            font-weight: 400 !important;
            line-height: 1.65 !important;
            padding: 13px 16px !important;
            transition: all 0.2s ease !important;
            caret-color: #c084fc !important;
        }

        textarea:hover, input[type="text"]:hover, input[type="number"]:hover {
            border-color: rgba(255,255,255,0.17) !important;
            background: rgba(255,255,255,0.065) !important;
        }

        textarea:focus, input[type="text"]:focus, input[type="number"]:focus {
            border-color: rgba(192,132,252,0.55) !important;
            background: rgba(192,132,252,0.05) !important;
            box-shadow: 0 0 0 4px rgba(192,132,252,0.1) !important;
            outline: none !important;
        }

        textarea::placeholder, input::placeholder {
            color: rgba(255,255,255,0.2) !important;
        }

        /* =============================================
           LABELS
        ============================================= */
        label {
            font-family: 'DM Sans', sans-serif !important;
            color: rgba(255,255,255,0.42) !important;
            font-size: 11px !important;
            font-weight: 600 !important;
            text-transform: uppercase !important;
            letter-spacing: 0.1em !important;
        }

        h2, h3 {
            font-family: 'Syne', sans-serif !important;
            color: rgba(255,255,255,0.9) !important;
            font-weight: 700 !important;
            font-size: 15px !important;
        }

        /* =============================================
           SELECT
        ============================================= */
        select {
            font-family: 'DM Sans', sans-serif !important;
            border-radius: 13px !important;
            border: 1px solid rgba(255,255,255,0.09) !important;
            background: rgba(255,255,255,0.045) !important;
            color: #f0f4ff !important;
            font-size: 14px !important;
            padding: 11px 14px !important;
            width: 100% !important;
        }

        select:focus {
            border-color: rgba(192,132,252,0.55) !important;
            box-shadow: 0 0 0 4px rgba(192,132,252,0.1) !important;
            outline: none !important;
        }

        /* =============================================
           SLIDERS
        ============================================= */
        input[type="range"] { accent-color: #c084fc !important; }

        /* =============================================
           ACCORDIONS
        ============================================= */
        .gr-accordion, details {
            border-radius: 18px !important;
            overflow: hidden !important;
            border: 1px solid rgba(255,255,255,0.07) !important;
            background: rgba(255,255,255,0.025) !important;
            margin-bottom: 10px !important;
        }

        summary {
            font-family: 'DM Sans', sans-serif !important;
            color: rgba(255,255,255,0.72) !important;
            font-size: 13px !important;
            font-weight: 600 !important;
            padding: 15px 20px !important;
            cursor: pointer !important;
            user-select: none;
            transition: all 0.2s ease;
        }

        summary:hover {
            background: rgba(255,255,255,0.04) !important;
            color: #fff !important;
        }

        /* =============================================
           OUTPUT IMAGE
        ============================================= */
        #output-image-box,
        #output-image-box *,
        #output-image-box > div,
        #output-image-box > div > div {
            height: auto !important;
            max-height: unset !important;
            min-height: 0 !important;
        }

        #output-image-box img {
            width: 100% !important;
            height: auto !important;
            max-height: unset !important;
            object-fit: contain !important;
            display: block !important;
            border-radius: 16px !important;
        }

        #output-image-box .image-frame,
        #output-image-box [data-testid="image"],
        #output-image-box [data-testid="image"] > div {
            height: auto !important;
            max-height: unset !important;
            background: transparent !important;
            border: none !important;
        }

        /* =============================================
           GALLERY
        ============================================= */
        #sd-gallery {
            height: auto !important;
            max-height: unset !important;
            min-height: 200px !important;
            width: 100% !important;
        }

        #sd-gallery > div,
        #sd-gallery > div > div,
        #sd-gallery [data-testid="gallery"],
        #sd-gallery [data-testid="gallery"] > div,
        #sd-gallery .grid-wrap,
        #sd-gallery .svelte-1pijsyv,
        #sd-gallery .svelte-hpz2c8 {
            height: auto !important;
            max-height: unset !important;
            min-height: 0 !important;
            overflow: visible !important;
        }

        #sd-gallery .grid-container,
        #sd-gallery ul,
        #sd-gallery ol {
            display: grid !important;
            grid-template-columns: repeat(5, 1fr) !important;
            gap: 10px !important;
            height: auto !important;
            max-height: unset !important;
            padding: 4px !important;
        }

        #sd-gallery li,
        #sd-gallery .thumbnail-item {
            height: auto !important;
            max-height: unset !important;
            border-radius: 12px !important;
            overflow: hidden !important;
            aspect-ratio: 1 / 1 !important;
            background: rgba(255,255,255,0.04) !important;
        }

        #sd-gallery li img,
        #sd-gallery .thumbnail-item img,
        #sd-gallery img {
            width: 100% !important;
            height: 100% !important;
            object-fit: cover !important;
            display: block !important;
            border-radius: 12px !important;
            transition: transform 0.25s ease, opacity 0.25s ease !important;
            opacity: 0.88;
        }

        #sd-gallery li img:hover,
        #sd-gallery .thumbnail-item img:hover,
        #sd-gallery img:hover {
            transform: scale(1.05) !important;
            opacity: 1 !important;
        }

        #gallery-box {
            background: rgba(255,255,255,0.032) !important;
            border-radius: 22px !important;
            border: 1px solid rgba(255,255,255,0.07) !important;
            padding: 22px !important;
            transition: border-color 0.25s ease !important;
        }

        #gallery-box:hover {
            border-color: rgba(192,132,252,0.2) !important;
        }

        /* =============================================
           CHECKBOX
        ============================================= */
        input[type="checkbox"] {
            accent-color: #c084fc !important;
            width: 16px !important;
            height: 16px !important;
        }

        /* =============================================
           READ-ONLY TEXTBOXES
        ============================================= */
        textarea[readonly],
        .gr-textbox-readonly textarea {
            background: rgba(0,0,0,0.28) !important;
            color: rgba(255,255,255,0.52) !important;
            font-family: 'SF Mono','Fira Code','Cascadia Code',monospace !important;
            font-size: 12px !important;
            line-height: 1.8 !important;
            border-color: rgba(255,255,255,0.05) !important;
        }

        /* =============================================
           EXAMPLES TABLE
        ============================================= */
        .gr-examples table {
            width: 100% !important;
            border-radius: 14px !important;
            overflow: hidden !important;
            border: 1px solid rgba(255,255,255,0.07) !important;
        }

        .gr-examples td, .gr-examples th {
            background: rgba(255,255,255,0.025) !important;
            color: rgba(255,255,255,0.5) !important;
            font-size: 12.5px !important;
            border-color: rgba(255,255,255,0.05) !important;
        }

        .gr-examples tr:hover td {
            background: rgba(192,132,252,0.07) !important;
            color: rgba(255,255,255,0.85) !important;
            cursor: pointer;
        }

        /* =============================================
           SITE FOOTER — using div.site-footer, NOT <footer>
        ============================================= */
        div.site-footer {
            display: block !important;
            visibility: visible !important;
            width: 100% !important;
            background: rgba(4,4,12,0.97) !important;
            border-top: 1px solid rgba(255,255,255,0.06) !important;
            padding: 52px 52px 36px !important;
            margin-top: 24px !important;
        }

        .footer-top {
            display: flex;
            justify-content: space-between;
            align-items: flex-start;
            flex-wrap: wrap;
            gap: 36px;
            padding-bottom: 40px;
            border-bottom: 1px solid rgba(255,255,255,0.06);
        }

        .footer-logo {
            font-family: 'Syne', sans-serif !important;
            font-size: 1.25rem;
            font-weight: 800;
            letter-spacing: -0.02em;
            background: linear-gradient(90deg, #c084fc, #f472b6);
            -webkit-background-clip: text;
            -webkit-text-fill-color: transparent;
            display: block;
            margin-bottom: 10px;
        }

        .footer-tagline {
            font-size: 13px;
            color: rgba(255,255,255,0.28);
            line-height: 1.65;
            max-width: 260px;
        }

        .footer-links {
            display: flex;
            gap: 56px;
            flex-wrap: wrap;
        }

        .footer-col-title {
            font-size: 10.5px;
            font-weight: 600;
            color: rgba(255,255,255,0.28);
            letter-spacing: 0.1em;
            text-transform: uppercase;
            margin-bottom: 14px;
        }

        .footer-col ul {
            list-style: none;
            padding: 0; margin: 0;
            display: flex;
            flex-direction: column;
            gap: 10px;
        }

        .footer-col ul li {
            font-size: 13px;
            color: rgba(255,255,255,0.38);
            cursor: default;
            transition: color 0.2s;
        }

        .footer-col ul li:hover { color: rgba(255,255,255,0.75); }

        .footer-bottom {
            display: flex;
            justify-content: space-between;
            align-items: center;
            flex-wrap: wrap;
            gap: 16px;
            padding-top: 28px;
        }

        .footer-copy {
            font-size: 12px;
            color: rgba(255,255,255,0.2);
        }

        .footer-built {
            display: flex;
            align-items: center;
            gap: 8px;
            font-size: 13.5px;
            font-weight: 500;
            color: rgba(255,255,255,0.35);
        }

        .footer-heart {
            color: #f472b6;
            animation: hb 1.6s ease-in-out infinite;
            display: inline-block;
        }

        @keyframes hb {
            0%,100% { transform: scale(1); }
            14%      { transform: scale(1.35); }
            28%      { transform: scale(1); }
            42%      { transform: scale(1.2); }
        }

        .built-name {
            background: linear-gradient(90deg, #c084fc, #f472b6);
            -webkit-background-clip: text;
            -webkit-text-fill-color: transparent;
            font-weight: 700;
            font-size: 14px;
        }

        .footer-badges {
            display: flex;
            gap: 8px;
            flex-wrap: wrap;
        }

        .fbadge {
            font-size: 11px;
            padding: 4px 13px;
            border-radius: 999px;
            border: 1px solid rgba(255,255,255,0.07);
            color: rgba(255,255,255,0.3);
            letter-spacing: 0.04em;
        }

        /* =============================================
           SCROLLBAR
        ============================================= */
        ::-webkit-scrollbar { width: 5px; height: 5px; }
        ::-webkit-scrollbar-track { background: transparent; }
        ::-webkit-scrollbar-thumb { background: rgba(192,132,252,0.22); border-radius: 99px; }
        ::-webkit-scrollbar-thumb:hover { background: rgba(192,132,252,0.45); }

        /* =============================================
           MOBILE
        ============================================= */
        @media (max-width: 768px) {
            .hero-title { font-size: 2.2rem; }
            .site-nav { padding: 0 20px; }
            .app-body { padding: 20px 16px 36px !important; }
            div.site-footer { padding: 36px 20px 24px !important; }
            .footer-links { gap: 28px; }
        }

        """

        # KEY FIX: JS now ONLY hides Gradio's native footer elements.
        # It no longer touches <footer> tags broadly — instead it targets
        # only specific Gradio svelte classes and API links.
        # Our custom footer uses <div class="site-footer"> so it's never caught.
        js_fix = """
        <script>
        (function() {

            /* ---------- helpers ---------- */
            function stripHeights(root) {
                if (!root) return;
                var els = root.querySelectorAll('*');
                els.forEach(function(el) {
                    if (el.style.height     && el.style.height     !== 'auto')   el.style.height     = 'auto';
                    if (el.style.maxHeight  && el.style.maxHeight  !== 'unset')  el.style.maxHeight  = 'unset';
                    if (el.style.minHeight  && el.style.minHeight  !== '0')      el.style.minHeight  = '0';
                    if (el.style.overflow   === 'hidden')                        el.style.overflow   = 'visible';
                });
                root.style.height    = 'auto';
                root.style.maxHeight = 'unset';
                root.style.overflow  = 'visible';
            }

            function hideGradioFooter() {
                /* ONLY hide Gradio's own footer bar elements.
                   We target svelte-hashed classes and API links specifically.
                   We do NOT touch any element with class "site-footer". */
                var gradioFooterSelectors = [
                    'a[href*="gradio.app"]',
                    'a[href*="huggingface.co/spaces"]',
                    'button[title="Settings"]',
                    'button[aria-label="Settings"]',
                    '.show-api',
                    '.built-with',
                    '.api-docs',
                    '.svelte-1lnpkp2',
                    '.svelte-mpyp5e',
                    '.svelte-1ax1toq'
                ].join(',');

                document.querySelectorAll(gradioFooterSelectors).forEach(function(el) {
                    /* Never hide our own footer */
                    if (el.closest('.site-footer')) return;
                    el.style.cssText = 'display:none!important;height:0!important;overflow:hidden!important;';
                });

                /* Hide the Gradio footer bar: a small bar at very bottom of page.
                   Only hide <footer> tags that are DIRECT children of body or
                   .gradio-container — never hide .site-footer divs. */
                document.querySelectorAll('body > footer, .gradio-container > footer').forEach(function(el) {
                    el.style.cssText = 'display:none!important;height:0!important;overflow:hidden!important;';
                });
            }

            function fixAll() {
                /* fix output image */
                stripHeights(document.getElementById('output-image-box'));

                /* fix gallery */
                var g = document.getElementById('sd-gallery');
                if (g) {
                    stripHeights(g);
                    if (g.parentElement) stripHeights(g.parentElement);
                }

                /* hide only Gradio's own footer bar */
                hideGradioFooter();

                /* Ensure our site-footer is always visible */
                document.querySelectorAll('div.site-footer').forEach(function(el) {
                    el.style.removeProperty('display');
                    el.style.removeProperty('visibility');
                    el.style.removeProperty('height');
                    el.style.removeProperty('overflow');
                });
            }

            /* Run on load + polling for first 8s */
            window.addEventListener('load', function() {
                fixAll();
                var n = 0;
                var t = setInterval(function() {
                    fixAll();
                    if (++n >= 16) clearInterval(t);
                }, 500);
            });

            /* MutationObserver */
            var observer = new MutationObserver(fixAll);

            document.addEventListener('DOMContentLoaded', function() {
                observer.observe(document.body, {
                    childList: true,
                    subtree: true,
                    attributes: true,
                    attributeFilter: ['style', 'class']
                });
                fixAll();
            });
        })();
        </script>
        """

        with gr.Blocks(
            title="Stable Diffusion Studio",
            theme=gr.themes.Base(
                primary_hue="violet",
                secondary_hue="pink",
                neutral_hue="slate",
                font=gr.themes.GoogleFont("DM Sans"),
            ),
            css=custom_css
        ) as interface:

            gr.HTML(js_fix)

            # ── NAVBAR ──
            gr.HTML("""
            <nav class="site-nav">
                <span class="nav-logo">✦ AR Studio</span>
                <div style="display:flex;align-items:center;gap:16px;">
                    <span style="font-size:12px;color:'white';
                                 letter-spacing:0.06em;text-transform:uppercase;
                                 font-family:'DM Sans',sans-serif;">
                        AI Image Generator
                    </span>
                </div>
            </nav>
            """)

            # ── HERO ──
            gr.HTML("""
            <div class="hero-section">
                <div class="hero-eyebrow">
                    <span class="hero-dot"></span>
                    Powered by Stable Diffusion
                </div>
                <h1 class="hero-title">
                    Create Stunning<br><span class="grad">AI Visuals</span>
                </h1>
                <p class="hero-sub">
                    Generate, explore, and refine AI&#8209;generated images with
                    professional&#8209;grade controls and a seamless workflow.
                </p>
                <div class="hero-stats">
                    <div class="stat-item">
                        <span class="stat-num">3</span>
                        <span class="stat-lbl">Models</span>
                    </div>
                    <div class="stat-item">
                        <span class="stat-num">5</span>
                        <span class="stat-lbl">Schedulers</span>
                    </div>
                    <div class="stat-item">
                        <span class="stat-num">∞</span>
                        <span class="stat-lbl">Creativity</span>
                    </div>
                </div>
            </div>
            """)

            gr.HTML('<div class="divider"></div>')

            # ── MAIN APP ──
            with gr.Row(elem_classes="app-body"):

                # LEFT MAIN COLUMN
                with gr.Column(scale=4):

                    with gr.Group(elem_classes="modern-card"):
                        gr.Markdown("## ✨ Create AI Art")

                        prompt = gr.Textbox(
                            label="🎯 Prompt",
                            placeholder="Describe your image... e.g. a cinematic futuristic city at golden hour, dramatic lighting, ultra-detailed...",
                            lines=4
                        )

                        negative_prompt = gr.Textbox(
                            label="🚫 Negative Prompt",
                            placeholder="What to avoid... e.g. blurry, low quality, distorted, bad anatomy...",
                            lines=2
                        )

                        generate_btn = gr.Button(
                            "✦  Generate Image",
                            variant="primary",
                            size="lg"
                        )

                    gr.Markdown("")

                    with gr.Row():

                        # OUTPUT IMAGE
                        with gr.Column(scale=2, elem_classes="output-img-card"):
                            output_image = gr.Image(
                                label="🖼️ Generated Output",
                                type="pil",
                                elem_id="output-image-box",
                                height=None,
                                show_download_button=True,
                                show_label=True
                            )

                        # DETAILS
                        with gr.Column(scale=1):
                            with gr.Group(elem_classes="modern-card"):
                                generation_info = gr.Textbox(
                                    label="📋 Generation Details",
                                    lines=14,
                                    interactive=False
                                )
                                saved_path = gr.Textbox(
                                    label="💾 Saved File Path",
                                    interactive=False
                                )

                    gr.Markdown("")

                    # GALLERY
                    with gr.Group(elem_id="gallery-box"):
                        gr.Markdown("## 🖼️ Gallery")
                        gallery = gr.Gallery(
                            value=[],
                            label="Generated Images",
                            show_label=False,
                            elem_id="sd-gallery",
                            columns=5,
                            rows=2,
                            object_fit="cover",
                            height="auto"
                        )

                # RIGHT SIDEBAR
                with gr.Column(scale=1, elem_classes="sidebar"):

                    with gr.Accordion("⚙️ Model Setup", open=True):

                        model_choice = gr.Dropdown(
                            choices=[
                                "Stable Diffusion 1.5 (Recommended)",
                                "Stable Diffusion 2.1",
                                "Realistic Vision (RealVisXL)"
                            ],
                            value="Stable Diffusion 1.5 (Recommended)",
                            label="Model Selection"
                        )

                        device_choice = gr.Dropdown(
                            choices=[
                                "Auto (Recommended)",
                                "GPU (CUDA)",
                                "CPU (Slower)"
                            ],
                            value="Auto (Recommended)",
                            label="Device Selection"
                        )

                        init_btn    = gr.Button("🚀  Initialize Model", variant="primary")
                        init_status = gr.Textbox(label="Initialization Status", lines=4)

                    with gr.Accordion("🎛️ Advanced Settings", open=False):

                        width    = gr.Slider(256, 1024, 512,  step=64,  label="Width")
                        height   = gr.Slider(256, 1024, 512,  step=64,  label="Height")
                        steps    = gr.Slider(10,  100,  20,   step=1,   label="Inference Steps")
                        guidance = gr.Slider(1.0, 20.0, 7.5,  step=0.5, label="Guidance Scale")

                        scheduler = gr.Dropdown(
                            choices=["euler_a","euler","ddim","dpm_solver","lms"],
                            value="euler_a",
                            label="Scheduler"
                        )

                        scheduler_info = gr.Textbox(
                            label="Scheduler Info",
                            interactive=False,
                            lines=2
                        )

                        seed       = gr.Number(-1,    label="Seed  (−1 = random)")
                        save_image = gr.Checkbox(True, label="💾 Auto-save Generated Images")

                    '''with gr.Accordion("📊 System Info", open=False):
                        memory_btn  = gr.Button("📊  Check Memory Usage")
                        memory_info = gr.Textbox(label="Memory Information", lines=8)'''

                    with gr.Accordion("✨ Example Prompts", open=False):
                        gr.Examples(
                            examples=self.get_example_prompts(),
                            inputs=[prompt, negative_prompt],
                            label="Click Any Example"
                        )

            # ── FOOTER ──
            # IMPORTANT: Using <div class="site-footer"> NOT <footer> tag
            # This prevents the JS from accidentally hiding it
            gr.HTML("""
            <div class="site-footer">
                <div class="footer-top">
                    <div>
                        <span class="footer-logo">✦ SD Studio</span>
                        <p class="footer-tagline">
                            A professional AI image generation workspace powered by
                            open&#8209;source Stable Diffusion models.
                        </p>
                    </div>
                    <div class="footer-links">
                        <div class="footer-col">
                            <div class="footer-col-title">Models</div>
                            <ul>
                                <li>SD 1.5</li>
                                <li>SD 2.1</li>
                                <li>RealVisXL</li>
                            </ul>
                        </div>
                        <div class="footer-col">
                            <div class="footer-col-title">Schedulers</div>
                            <ul>
                                <li>Euler A</li>
                                <li>DDIM</li>
                                <li>DPM Solver</li>
                                <li>LMS</li>
                            </ul>
                        </div>
                        <div class="footer-col">
                            <div class="footer-col-title">Tech Stack</div>
                            <ul>
                                <li>🤗 Diffusers</li>
                                <li>🔥 PyTorch</li>
                                <li>🎨 Gradio</li>
                                <li>🐍 Python</li>
                            </ul>
                        </div>
                    </div>
                </div>
                <div class="footer-bottom">
                    <span class="footer-copy">
                        © 2025 Stable Diffusion Studio &nbsp;·&nbsp; All rights reserved
                    </span>
                    <div class="footer-badges">
                        <span class="fbadge">Gradio</span>
                        <span class="fbadge">PyTorch</span>
                        <span class="fbadge">Diffusers</span>
                        <span class="fbadge">CUDA</span>
                    </div>
                    <div class="footer-built">
                        <span class="footer-heart">♥</span>
                        Built with passion by
                        <span class="built-name">MANJU A R</span>
                    </div>
                </div>
            </div>
            """)

            # ── EVENTS ──
            init_btn.click(
                fn=self.initialize_generator,
                inputs=[model_choice, device_choice],
                outputs=init_status
            )

            generate_btn.click(
                fn=self.generate_image,
                inputs=[
                    prompt, negative_prompt,
                    width, height, steps, guidance,
                    scheduler, seed, save_image
                ],
                outputs=[output_image, generation_info, saved_path]
            ).then(
                fn=lambda: self.gallery_images,
                outputs=gallery
            )

            scheduler.change(
                fn=self.show_scheduler_info,
                inputs=scheduler,
                outputs=scheduler_info
            )

            '''memory_btn.click(
                fn=self.get_memory_info,
                outputs=memory_info
            )'''

        return interface


```
